### Q1)  A F&B manager wants to determine whether there is any
### significant difference in the diameter of the cutlet between two
### units. A randomly selected sample of cutlets was collected from
### both units and measured? Analyze the data and draw inferences
### at 5% significance level. Please state the assumptions and tests
### that you carried out to check validity of the assumptions.

```python
import pandas as pd
import numpy as np
from scipy import stats
```

```python
ls
```

```python
help(pd.read_csv)
```

```python
Cutlets = pd.read_csv('D:\Hypothesis CSV assignment\Cutlets.csv')
```

```python
Cutlets
```

#### Hypothesis Formulation

H0: There is no significant difference in diameter of the cutlet between two units.# u1 = u2
Ha: There is a significant difference in diameter of the cutlet between two units.#u1 != u2 #### Level of signficance = 5%
##### 2 sample independent T test

```python
t_val,p_val = stats.ttest_ind(a = Cutlets['Unit A'], b = Cutlets['Unit B'])
t_val,p_val
```

```python
p_val = p_val/2
p_val
```

```python
if p_val<0.05:
    print('we reject the null hypothesis and we can say that there is a significant difference in diameter of the cutlets between two units')
else:
    print('we do not reject the null hypothesis and we can say that there is no signficant difference in diameter of the cutlets between two units')
```

### Q2) A hospital wants to determine whether there is any
### difference in the average Turn Around Time (TAT) of reports
### of the laboratories on their preferred list. They collected a
### random sample and recorded TAT for reports of 4
### laboratories. TAT is defined as sample collected to report
### dispatch.

### Analyze the data and determine whether there is any
### difference in average TAT among the different laboratories
### at 5% significance level.

```python
LabTAT = pd.read_csv('D:\Hypothesis CSV assignment\LabTAT.csv')
```

```python
LabTAT
```

#### Hypothesis Formulation:

H0: All the average TAT of reports of the laboratories are equal.
Ha: At least one average TAT of reports of laboratories is different.#### Level of significance = 5% #### We perform ANOVA Test.

```python
statistics,p_val = stats.f_oneway(LabTAT['Laboratory 1'],LabTAT['Laboratory 2'],LabTAT['Laboratory 3'],LabTAT['Laboratory 4'])
statistics,p_val
```

```python
if p_val<0.05:
    print('We reject the null hypothesis and we can say that there is signficant difference in average TAT of reports of the laboratories.')
else:
    print('We do not reject the null hypothesis and we can say that there is no signficant difference in average TAT of reports of the laboratories.')
```

### Q3) Sales of products in four different regions is tabulated for males and females.
### Find if male-female buyer rations are similar across regions.



```python
BuyerRatio = pd.read_csv('D:\Hypothesis CSV assignment\BuyerRatio.csv')
```

```python
BuyerRatio
```

#### Hypothesis Formulation

H0:All proportions are equal.
Ha: Not all proportion are equal. #### Level of signficance = 5% (Considered as 5% as level of signficance is not mentioned)
We perform Chi-squared Test

```python
del BuyerRatio['Observed Values']
```

```python
BuyerRatio
```

```python
stats.chi2_contingency(observed = BuyerRatio)
```

```python
chi2_score,p_val,dof,expected = stats.chi2_contingency(observed = BuyerRatio)
print('***************************************************')
print('CHI-SQAURE TEST RESULTS')
print('CHI2 Value           :',chi2_score)
print('P-value              :',p_val)
print('Degree of Freedom    :',dof)
print('Expected Table       :\n',expected)
print('***************************************************')
```

```python
if p_val<0.05:
    print('We reject the null hypothesis and we can say that Not all proportion are equal.')
else:
    print('We do not reject the null hypothesis and we can say that all proportion are equal.')
```

### customer order forms. They audit a certain % of the

### at 5% significance level and help the manager draw

```python
Customer = pd.read_csv('D:\Hypothesis CSV assignment\Costomer+OrderForm.csv')
```

```python
Customer_values = pd.DataFrame({'Phillippines':[271,29],'Indonesia':[267,33],'Malta':[269,31],'India':[280,20]})
Customer_values
```

#### Hypothesis Formulation:

H0: The defective percentage does not varies significantly by centre.
Ha: The defective percentage does varies significantly by centre. ##### Level of signficance = 5% #### We perform Chi-squared Test.

```python
stats.chi2_contingency(observed = Customer_values)
```

```python
chi2_score,p_val,dof,expected = stats.chi2_contingency(observed = Customer_values)
print('***************************************************')
print('CHI-SQAURE TEST RESULTS')
print('CHI2 Value           :',chi2_score)
print('P-value              :',p_val)
print('Degree of Freedom    :',dof)
print('Expected Table       :\n',expected)
print('***************************************************')
```

```python
if p_val<0.05:
    print('We reject the null hypothesis and we can say that the defective percentage does not varies significantly by centre.')
else:
    print('We do not reject the null hypothesis and we can say that the defective percentage does not varies significanly by centre.')
```
